```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
kp2020 = pd.read_csv('data/KP2020.csv', encoding='cp949')
print(kp2020.shape)
display(kp2020.head())
```

```python
kp2021 = pd.read_csv('data/KP2021.csv', encoding='cp949')
print(kp2021.shape)
display(kp2021.head())
```

```python
npa2020 = pd.read_csv('data/NPA2020.csv', encoding='cp949')
print(npa2020.shape)
display(npa2020.head())
```

## kp 보이스피싱 데이터 병합

```python
kp215_2020 = kp2020[kp2020['EVT_CL_CD'] == 215]
kp215_2020.head(2)
```

```python
kp215_2021 =  kp2021[kp2021['EVT_CL_CD'] == 215]
kp215_2021.head(2)
```

```python
kp215 = pd.concat([kp215_2020, kp215_2021], ignore_index=True)
kp215.head()
```

## kp 데이터 전처리

```python
kp215['RECV_CPLT_DM'] = kp215['RECV_CPLT_DM'].apply(lambda x: pd.to_datetime(x.split('.')[0], format='%y/%m/%d %H:%M:%S'))
```

```python
kp215.head()
```

## npa 데이터 전처리

```python
npa215 = npa2020[npa2020['EVT_CL_CD']==215]
npa215.head(2)
```

```python
npa215.insert(0, 'RECV_DEPT_NM', np.nan)
npa215.loc[npa215['NPA_CL'] == 13, 'RECV_DEPT_NM'] = '대전청'
npa215.loc[npa215['NPA_CL'] == 19, 'RECV_DEPT_NM'] = '충남청'
```

```python
npa215.head(2)
```

```python
npa215['RECV_CPLT_TM'] = npa215['RECV_CPLT_TM'].astype('str')
npa215['RECV_CPLT_TM'] = npa215['RECV_CPLT_TM'].str.rjust(width=6, fillchar='0')
```

```python
npa215['RECV_CPLT_DM'] = npa215['RECV_CPLT_DT'].astype('str') + npa215['RECV_CPLT_TM']
```

```python
npa215['RECV_CPLT_DM'] = pd.to_datetime(npa215['RECV_CPLT_DM'])
```

```python
npa215.drop('RECV_CPLT_DT', axis=1, inplace=True)
npa215.drop('RECV_CPLT_TM', axis=1, inplace=True)
npa215 = npa215.rename(columns={'HPPN_OLD_ADDR':'HPPN_PNU_ADDR'},)
```

## kp, npa 병합

```python
kpa215 = pd.concat([kp215, npa215], ignore_index=True)
kpa215.head(2)
```

```python
kpa215.drop('EVT_CL_CD',axis=1, inplace=True)
kpa215 = kpa215.rename(columns={'RECV_DEPT_NM':'접수부서', 'RECV_CPLT_DM':'접수완료일시', 'NPA_CL':'경찰청구분', 'EVT_STAT_CD':'사건상태', 'RPTER_SEX':'성별', 'HPPN_PNU_ADDR':'발생지점','HPPN_X':'발생좌표X', 'HPPN_Y':'발생좌표Y', 'SME_EVT_YN':'동일사건여부'})
```

## kpa 변수명 변경

```python
codebook = pd.read_excel('data/codeBook_v3.xlsx')
```

```python
npa_cl = dict(zip(codebook.loc[codebook['컬럼명'] == 'NPA_CL']['코드값'].astype(str).tolist(), codebook.loc[codebook['컬럼명'] == 'NPA_CL']['코드명'].tolist()))
```

```python
kpa215['경찰청구분'] = kpa215['경찰청구분'].astype('str').map(npa_cl)
```

```python
evt_st = dict(zip(codebook.loc[codebook['컬럼명'] == 'EVT_STAT_CD']['코드값'].astype(str).tolist(), codebook.loc[codebook['컬럼명'] == 'EVT_STAT_CD']['코드명'].tolist()))
```

```python
kpa215['사건상태'] = kpa215['사건상태'].astype('str').map(evt_st)
```

```python
def sex(n):
    if n == 1:
        return '남성'
    elif n == 2:
        return '여성'
    else:
        return '불상'
```

```python
kpa215['성별'] = kpa215['성별'].map(sex)
```

## 데이터 저장

```python
kpa215.to_csv('data/kpa215.csv', index=False)
```

### 데이터 정리
- kp는 대전 세종 충남 지역의 사건들을 모아놨다.('충남청', '대전청', '2지구대(충남)', '남대전지구대', '대전대덕', '대전동부', '대전파출소', '대전역지구대', '대전중부', '대전둔산', '대전서부', '서대전지구대', '대전유성', '세종로파출소', '세종', '세종청')
- npa는 대전청과 충남청 사건들만 모아놨다.
- 취급하는 사건의 갯수는 npa 데이터가 더 많다.
- 보이스 피싱 사건만 두고 봤을 때도 npa 데이터가 훨씬 많다 왜?
- 시간순 정렬
- npa는 2020년 1월\~11월 / kp는 2020년 12월, 2021년 3월\~ 2023년 1월
